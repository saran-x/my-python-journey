```python
from typing import List,TypeVar,Dict

# Declaration of the T
T = TypeVar('T')
# generics
def get_first_item[T](items: List[T]) -> T:
    return items

listss: List[int] = [1,2,3,4]
args: int = get_first_item(items= listss)
print(args)
listss: List[str] = ['a','b','c','d']
args: str = get_first_item(items= listss)
print(args)

```

```python
# Create a simple program that takes a pizza order and calculates the total cost. ---> using generics 
# custom exception
class Invalid_input(Exception):
    """ Invalid input plese enter a correct value"""
# Pizza calculation function
def pizza_calculation[str,float](size:str,pep:str,chees:str,delv:str) -> float:
    p_size = 8.00  if size == 's'   else ( 11.99 if size == 'm' else 14.99 )
    pep_am = 0     if pep == 'n'    else ( 2.00 if size == 's' else 3.00 )
    extra_ch = 0   if chees == 'n'  else 1.50
    delivery = 0   if delv == 'n'   else 5.00
    total = p_size+pep_am+extra_ch+delivery
    print("--- Your Order ---")
    print(f"Size: {"Small"  if size == 's' else ('Medium' if size == 'm' else 'Large')} ${p_size}")
    print(f"Pepperoni: ${pep_am}")
    print(f"Extra Cheese: ${extra_ch}")
    print(f"Delivery : ${delivery}")
    print("------------------")
    return total


# Input condition
valid_sizes = ['s','m','l']
valid_yn = ['y','n']
try:
    piza_size = input("What size pizza? (S/M/L): ").lower()
    if piza_size not in valid_sizes:
        raise Invalid_input("Invalid format only under (S/M/L)")
    
    pepperoni = input("Do you want pepperoni? (Y/N): ").lower()
    if pepperoni not in valid_yn:
        raise Invalid_input("Invalid format only under (Y/N)")
    
    cheeses = input("Do you want extra cheese? (Y/N): ").lower()
    if cheeses not in valid_yn:
        raise Invalid_input("Invalid format only under (Y/N)")
    
    delivery = input("Door Delivery? (Y/N): ").lower()
    if delivery not in valid_yn:
        raise Invalid_input("Invalid format only under (Y/N)")
    
    # Calculate and display
    total = pizza_calculation(
        size=piza_size, 
        pep=pepperoni, 
        chees=cheeses, 
        delv=delivery
    )
    print(f"TOTAL: ${total:.2f}")
    print("Thank you for ordering!")

except Invalid_input as e:
    print(f"Error: {e}")
```
